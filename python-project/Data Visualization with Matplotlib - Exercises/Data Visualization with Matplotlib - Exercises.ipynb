```python
import numpy as np
x = np.arange(0,100)
y = x*2
z = x**2
```

**Import matplotlib.pyplot as plt and use the command %matplotlib inline if using a jupyter notebook. If not using a jupyter notebook, what other command can be used instead of %matplotlib inline?**

```python
import matplotlib.pyplot as plt
```

## Exercise 1

**Follow these steps:**

* **Create a figure object called 'fig' using plt.figure().**
* **Add a figure canvas using add_axes, with a list of [0,0,1,1]. Update the axis with the variable 'ax'.**
* **Plot (x,y) on the axes and add labels and titles to the plot.**

```python
fig = plt.figure(figsize=(2,2))
ax = fig.add_axes([0,0,1,1])
ax.plot(x, y)
ax.set_xlabel('x label')
ax.set_ylabel('y label')
ax.set_title('x and y')
```

## Exercise 2
**Create a figure and two axes, setting the range of both axes, ax1 and ax2, to [0,0,1,1] and [0.2,0.5,.2,.2], respectively, then, plot (x,y) on both axes and display the figure.**

```python
# create canvas
fig = plt.figure(dpi=72)

# main
ax1 = fig.add_axes([0,0,1,1])

# inset
ax2 = fig.add_axes([0.2,0.5,.2,.2])

# main plot
ax1.plot(x, y)
ax1.set_xlabel('x label')
ax1.set_ylabel('y label')
ax1.set_title('Main')


# inset plot
ax2.plot(x, y, 'r')
ax2.set_xlabel('x label')
ax2.set_ylabel('y label')
ax2.set_title('Sub')
```

## Exercise 3
**Plot using two sets of axes, setting the range of both axes to [0,0,1,1] and [0.2,0.5,.4,.4], respectively. Then, use the arrays x, y, and z to plot on range [20, 22] in x and [30, 50] in y**


```python
fig = plt.figure(dpi=72)
ax1 = fig.add_axes([0, 0, 1, 1])
ax2 = fig.add_axes([0.2, 0.5, 0.4, 0.4])

# main plot
ax1.plot(x, z)
ax1.set_xlabel('x label')
ax1.set_ylabel('z label')
ax1.set_title('x - z')

# inset plot
ax2.plot(x, y, 'r')
ax2.set_xlabel('x label')
ax2.set_ylabel('y label')
ax2.set_title('x - y zoom')
ax2.set_xlim([20, 22])
ax2.set_ylim([30, 50])
```

## Exercise 4

**Create a subplot with 1 row and 2 columns to plot two sets of data. Then, plot (x,y) and (x,z) on the axes of the created canvas, and use the linewidth and style commands to decorate the lines of the graph.**

```python
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(14, 4), dpi=80)

# supplot 1
ax1.plot(x, y, 'b--')
ax1.set_xlabel('x label')
ax1.set_ylabel('y label')
ax1.set_title('x - y')

# supplot 2
ax2.plot(x, z, 'r')
ax2.set_xlabel('x label')
ax2.set_ylabel('z label')
ax2.set_title('x - z')
```
